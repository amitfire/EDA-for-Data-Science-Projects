```
ser_obj  =pd.Series(['hello', 'how' ,'are', 'you'])
```

```
ser_obj
```

```
for i in ser_obj:   
   i.replace(i[0],i[0].upper())
ser_obj
```

```
dict_len = {}
for i in ser_obj:
  print(f'length of {i}: {len(i)}')
  dict_len[i]=len(i)

print('Lengthiest Word is',list(dict_len.keys())[list(dict_len.values()).index(max(list(dict_len.values())))])
```
